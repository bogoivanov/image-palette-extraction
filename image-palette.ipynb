```python
import numpy as np
import matplotlib.pyplot as plt

def plot_color_wheel():
    # Create theta for 256 angles between 0 and 2*pi
    theta = np.linspace(0, 2.*np.pi, 256, endpoint=True)

    # Create the color wheel
    fig, ax = plt.subplots(figsize=(12, 12), subplot_kw=dict(projection='polar'))

    # Create a colormap
    cmap = plt.get_cmap("hsv")

    # Plot the color wheel
    radii = np.array([1 for _ in theta])
    bars = ax.bar(theta, radii, width=theta[1] - theta[0], bottom=0.0, color=cmap(theta / (2.*np.pi)))

    # Hide labels and ticks
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.set_xticks([])
    ax.set_yticks([])

    # Show the plot
    plt.show()

# To plot the color wheel, just call the function
plot_color_wheel()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def create_and_plot_color_square(color, size=50):
    """
    Create and plot a square of a specific color.
    
    Parameters:
    - color (tuple): RGB values in the range [0, 255], e.g., (255, 0, 0) for red.
    - size (int): Size of the square in pixels (default is 50).
    """
    
    # Scale the color values to the range [0, 1]
    scaled_color = tuple([c/255. for c in color])
    
    # Create an array of the specified color
    square = np.ones((size, size, 3))
    square[:, :] = scaled_color

    # Plot the square
    fig, ax = plt.subplots()
    ax.imshow(square)
    
    # Remove axis ticks and labels
    ax.axis('off')
    
    plt.show()
colour = [255, 240, 96]
# To create and plot a color square, call the function with RGB values in the range [0, 255]:
create_and_plot_color_square(colour)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def complementary_color(rgb_color):
    """
    Compute the complementary color for a given RGB color.
    
    Parameters:
    - rgb_color (tuple): Original RGB values in the range [0, 255].
    
    Returns:
    - tuple: Complementary RGB values in the range [0, 255].
    """
    return tuple(255 - c for c in rgb_color)

def create_and_plot_complementary_square(rgb_color, size=50):
    """
    Create and plot a square of the complementary color.
    
    Parameters:
    - rgb_color (tuple): Original RGB values in the range [0, 255].
    - size (int): Size of the square in pixels (default is 50).
    """
    
    # Calculate the complementary color
    comp_color = complementary_color(rgb_color)
    
    # Scale the complementary color values to the range [0, 1]
    scaled_color = tuple([c/255. for c in comp_color])
    
    # Create an array of the complementary color
    square = np.ones((size, size, 3))
    square[:, :] = scaled_color

    # Plot the square
    fig, ax = plt.subplots()
    ax.imshow(square)
    
    # Remove axis ticks and labels
    ax.axis('off')
    
    plt.show()
colour = [255, 240, 96]

# To create and plot a square of the complementary color for red:
create_and_plot_complementary_square(colour)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_soccer_field():
    # Field dimensions (in meters)
    field_length = 105.0
    field_width = 68.0

    # Define the figure and axis
    fig, ax = plt.subplots(figsize=(10, 7))

    # Set the background color as green
    ax.set_facecolor('green')
    
    # Field border
    border = patches.Rectangle((0,0), field_length, field_width, edgecolor="white", facecolor="none", lw=2)
    ax.add_patch(border)
    
    # Center line
    ax.plot([field_length/2, field_length/2], [0, field_width], color="white")
    
    # Goals
    ax.plot([0,0], [field_width/2 - 7.32/2, field_width/2 + 7.32/2], color="white")
    ax.plot([field_length, field_length], [field_width/2 - 7.32/2, field_width/2 + 7.32/2], color="white")
    
    # Center circle
    center_circle = patches.Circle((field_length/2, field_width/2), 9.15, color="white", fill=False)
    ax.add_patch(center_circle)

    # Center spot
    center_spot = patches.Circle((field_length/2, field_width/2), 0.37, color="white")
    ax.add_patch(center_spot)
    
    # Penalty areas
    left_penalty_area = patches.Rectangle((0, (field_width/2) - 16.5), 16.5, 33, edgecolor="white", facecolor="none")
    right_penalty_area = patches.Rectangle((field_length - 16.5, (field_width/2) - 16.5), 16.5, 33, edgecolor="white", facecolor="none")
    ax.add_patch(left_penalty_area)
    ax.add_patch(right_penalty_area)
    
    # Goal areas
    left_goal_area = patches.Rectangle((0, (field_width/2) - 5.5), 5.5, 11, edgecolor="white", facecolor="none")
    right_goal_area = patches.Rectangle((field_length - 5.5, (field_width/2) - 5.5), 5.5, 11, edgecolor="white", facecolor="none")
    ax.add_patch(left_goal_area)
    ax.add_patch(right_goal_area)

    # Penalty spots
    left_penalty_spot = patches.Circle((11, field_width/2), 0.37, color="white")
    right_penalty_spot = patches.Circle((field_length - 11, field_width/2), 0.37, color="white")
    ax.add_patch(left_penalty_spot)
    ax.add_patch(right_penalty_spot)

    # "D" Arcs in front of goal fields
    left_d_arc = patches.Arc((11, field_width/2), 18.3, 18.3, angle=0, theta1=308, theta2=52, color="white")
    right_d_arc = patches.Arc((field_length - 11, field_width/2), 18.3, 18.3, angle=0, theta1=128, theta2=232, color="white")
    ax.add_patch(left_d_arc)
    ax.add_patch(right_d_arc)
    
    # Display settings
    ax.set_xlim(0, field_length)
    ax.set_ylim(0, field_width)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title("Soccer Field")

    plt.show()

plot_soccer_field()
```

```python
def count_files_in_directory(folder_path, extension=None):
    """Count the number of files in a directory with a specific extension."""
    if extension:
        return len([f for f in os.listdir(folder_path) if f.endswith(f'.{extension}')])
    else:
        return len([f for f in os.listdir(folder_path)])

# Example usage:
folder_path = 'original_images'
number_files_in_directory = count_files_in_directory(folder_path, extension='jpg')
print(f"Number of JPG files in the directory: {number_files_in_directory}")
```

```python
def get_image_names():
    image_paths = []
    for i in range(1, number_files_in_directory+1):
        image_path = f'original_images/IMAGE-{i}.jpg'
        image_paths.append(image_path)
    return image_paths
```

```python
path_to_your_images = get_image_names()
constant_height = 500
n_clusters = 5
```

```python
def rgb_to_cmyk(r, g, b):
    r /= 255.0
    g /= 255.0
    b /= 255.0
    k = 1 - max(r, g, b)
    if k == 1:
        c = m = y = 0
    else:
        c = (1 - r - k) / (1 - k)
        m = (1 - g - k) / (1 - k)
        y = (1 - b - k) / (1 - k)
    return c, m, y, k



def load_and_preprocess_image(path, scale_factor=2.0):
    try:
        image = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB)
        new_dimensions = (int(image.shape[1] * scale_factor), int(image.shape[0] * scale_factor))
        image = cv2.resize(image, new_dimensions, interpolation=cv2.INTER_LANCZOS4)
        image = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
        
        
          # You can use cv2.INTER_CUBIC for better quality
        # INTER_NEAREST: int
        # INTER_LINEAR: int
        # INTER_CUBIC: int
        # INTER_AREA: int
        # INTER_LANCZOS4: int
        # INTER_LINEAR_EXACT: int
        # INTER_NEAREST_EXACT: int
        # INTER_MAX: int
        # WARP_FILL_OUTLIERS: int
        # WARP_INVERSE_MAP: int
        # InterpolationFlags = int
        # """One of [INTER_NEAREST, INTER_LINEAR, INTER_CUBIC, INTER_AREA, INTER_LANCZOS4, INTER_LINEAR_EXACT, INTER_NEAREST_EXACT, INTER_MAX, WARP_FILL_OUTLIERS, WARP_INVERSE_MAP]"""

        return image
    except Exception as e:
        print(f"Failed to load {path}: {e}")
        return None
    




def apply_kmeans(pixels, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, n_init=10)
    labels = kmeans.fit_predict(pixels)
    centers = kmeans.cluster_centers_.astype(int)
    return centers

def create_palette(centers):
    palette = np.zeros((50, len(centers) * 50, 3), dtype=np.uint8)
    for i, color in enumerate(centers):
        palette[:, i * 50:(i + 1) * 50, :] = color
    return palette

def combine_images(image, palette, border_width=20):
    white_border = np.ones((image.shape[0], border_width, 3), dtype=np.uint8) * 255
    return np.concatenate((image, white_border, palette), axis=1)

def print_cmyk_values(centers):
    for i, color in enumerate(centers):
        c, m, y, k = rgb_to_cmyk(color[0], color[1], color[2])
        print(f"Color {i+1} - CMYK: (C:{c:.2f}, M:{m:.2f}, Y:{y:.2f}, K:{k:.2f})")

# Main code
def save_image(path, combined):
    cv2.imwrite(path, cv2.cvtColor(combined, cv2.COLOR_RGB2BGR))
    
# Main code
def plot_combined_image(image, title):
    plt.figure(figsize=(12, 6))
    plt.imshow(image)
    plt.axis('off')
    plt.title(title)
    plt.tight_layout()
    plt.show()

def generate_title(index, n_clusters):
    return f'Original image and extracted palette with 5 colors - {index + 1}'
```

```python
# Main code


for idx, path in enumerate(path_to_your_images):
    image = load_and_preprocess_image(path)
    if image is None:
        continue
    pixels = image.reshape(-1, 3)
    centers = apply_kmeans(pixels, n_clusters)
    palette = create_palette(centers)
    palette_resized = cv2.resize(palette, (palette.shape[1], image.shape[0]))
    combined = combine_images(image, palette_resized)
    title = generate_title(idx, n_clusters)
    current_time = datetime.now().strftime("%Y%m%d%H%M%S")
    save_image(f'output_images/resized_combined_image_{current_time}.png', combined)
    print_cmyk_values(centers)
    plot_combined_image(combined, title)
```

```python
import cv2
import numpy as np

# Read source image.
im_src = cv2.imread('books/book2.jpg')

# Four corners of the book in source image
pts_src = np.array([[141, 131], [480, 159], [493, 630], [64, 601]])

# Read destination image.
im_dst = cv2.imread('books/book1.jpg')

# Four corners of the book in destination image.
pts_dst = np.array([[318, 256], [534, 372], [316, 670], [73, 473]])

# Calculate Homography
h, status = cv2.findHomography(pts_src, pts_dst)

# Warp source image to destination based on homography
im_out = cv2.warpPerspective(im_src, h, (im_dst.shape[1], im_dst.shape[0]))

# Display images
cv2.imshow("Source Image", im_src)
cv2.imshow("Destination Image", im_dst)
cv2.imshow("Warped Source Image", im_out)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
